```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
attributes_container = []

# Using TwitterSearchScraper to scrape data and append tweets to list

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('FIFAWorldcup2022 since:2022-08-19 until:2022-09-20').get_items()):
    if i>500:
        break
    attributes_container.append([tweet.date, tweet.content])
    
# Creating a dataframe to load the list
tweets_df1 = pd.DataFrame(attributes_container, columns=["Date Created", "Tweet"])

tweets_df1
```
